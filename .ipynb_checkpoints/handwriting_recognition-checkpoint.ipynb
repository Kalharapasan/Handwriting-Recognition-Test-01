```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image, ImageDraw
import ipywidgets as widgets

from IPython.display import display, clear_output
import sklearn
from sklearn.model_selection import train_test_split
import oss
```

```python
print("TensorFlow version:", tf.__version__)
print("All libraries imported successfully!")
```

```python
# 1. Data Loading and Preprocessing
def load_and_preprocess_data():
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0
    x_train = x_train.reshape(-1, 28, 28, 1)
    x_test = x_test.reshape(-1, 28, 28, 1)
    return (x_train, y_train), (x_test, y_test)
```

```python
(x_train, y_train), (x_test, y_test) = load_and_preprocess_data()
print(f"Training data shape: {x_train.shape}")
print(f"Training labels shape: {y_train.shape}")
print(f"Test data shape: {x_test.shape}")
print(f"Test labels shape: {y_test.shape}")
```

```python
#2. Data Visualization
def visualize_sample_data(x_data, y_data, num_samples=10):
    fig, axes = plt.subplots(2, 5, figsize=(12, 6))
    axes = axes.ravel()
    
    for i in range(num_samples):
        axes[i].imshow(x_data[i].reshape(28, 28), cmap='gray')
        axes[i].set_title(f"Label: {np.argmax(y_data[i])}")
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

visualize_sample_data(x_train, y_train)
```

```python
# 3. Neural Network Model Architecture
def create_cnn_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.25),
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    return model
```

```python
model = create_cnn_model()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.summary()
```

```python
# 4. Model Training
callbacks = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(factor=0.2, patience=3)
]
```

```python
print("Starting model training...")
history = model.fit(
    x_train, y_train,
    batch_size=128,
    epochs=50,
    validation_data=(x_test, y_test),
    callbacks=callbacks,
    verbose=1
)
```

```python
# 5. Model Evaluation
def evaluate_model(model, x_test, y_test):
    test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
    
    print(f"Test Accuracy: {test_accuracy:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)
    
    return test_accuracy, test_loss, y_pred_classes, y_true_classes
```

```python
test_accuracy, test_loss, y_pred, y_true = evaluate_model(model, x_test, y_test)
```

```python
#6. Training Visualization
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    ax1.plot(history.history['accuracy'], label='Training Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(True)
    ax2.plot(history.history['loss'], label='Training Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()
    
plot_training_history(history)
```

```python
# 7. Interactive Drawing Interface
class DrawingCanvas:
    def __init__(self, width=280, height=280):
        self.width = width
        self.height = height
        self.image = Image.new('L', (width, height), 0)
        self.draw = ImageDraw.Draw(self.image)
        self.brush_size = 15

    def draw_point(self, x, y):
        self.draw.ellipse([x-self.brush_size, y-self.brush_size, 
                          x+self.brush_size, y+self.brush_size], fill=255)

    def clear(self):
        self.image = Image.new('L', (self.width, self.height), 0)
        self.draw = ImageDraw.Draw(self.image)

    def get_image_array(self):
        return np.array(self.image)

    def preprocess_for_prediction(self):
        img = self.image.resize((28, 28), Image.Resampling.LANCZOS)
        img_array = np.array(img)
        img_array = img_array.astype('float32') / 255.0
        img_array = img_array.reshape(1, 28, 28, 1)
        return img_array
```

```python
# 8. Real-time Prediction Interface
canvas_output = widgets.Output()
prediction_output = widgets.Output()
clear_button = widgets.Button(description="Clear Canvas")
predict_button = widgets.Button(description="Predict")
save_model_button = widgets.Button(description="Save Model")
load_model_button = widgets.Button(description="Load Model")
```

```python
drawing_canvas = DrawingCanvas()
```

```python
def setup_interactive_ui():
    button_layout = widgets.HBox([clear_button, predict_button, save_model_button, load_model_button])
    main_layout = widgets.VBox([
        widgets.Label("Draw a digit (0-9) in the canvas below:"),
        canvas_output,
        button_layout,
        prediction_output
    ])
    
    return main_layout
```

```python
def handle_canvas_interaction(x, y):
    with canvas_output:
        drawing_canvas.draw_point(x, y)
        display_canvas()
```

```python
def display_canvas():
    canvas_output.clear_output(wait=True)
    with canvas_output:
        plt.figure(figsize=(6, 6))
        plt.imshow(drawing_canvas.get_image_array(), cmap='gray')
        plt.title("Your Drawing")
        plt.axis('off')
        plt.show()
```

```python
def on_clear_button_clicked(b):
    drawing_canvas.clear()
    prediction_output.clear_output()
    display_canvas()
```

```python
def on_predict_button_clicked(b):
    processed_image = drawing_canvas.preprocess_for_prediction()
    prediction = model.predict(processed_image, verbose=0)
    predicted_digit = np.argmax(prediction)
    confidence = np.max(prediction)
    with prediction_output:
        prediction_output.clear_output()
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        ax1.imshow(processed_image.reshape(28, 28), cmap='gray')
        ax1.set_title(f"Processed Image\nPredicted: {predicted_digit}")
        ax1.axis('off')
        digits = range(10)
        ax2.bar(digits, prediction[0])
        ax2.set_xlabel('Digits')
        ax2.set_ylabel('Probability')
        ax2.set_title('Prediction Probabilities')
        ax2.set_xticks(digits)

        ax2.bar(predicted_digit, prediction[0][predicted_digit], color='red', alpha=0.7)
        
        plt.tight_layout()
        plt.show()
        
        print(f"Prediction: {predicted_digit}")
        print(f"Confidence: {confidence:.2%}")
```

```python
def on_save_model_clicked(b):
    model.save('models/handwriting_model.h5')
    print("Model saved successfully!")
```

```python
def on_load_model_clicked(b):
    global model
    try:
        model = keras.models.load_model('models/handwriting_model.h5')
        print("Model loaded successfully!")
    except:
        print("No saved model found!")
```

```python
clear_button.on_click(on_clear_button_clicked)
predict_button.on_click(on_predict_button_clicked)
save_model_button.on_click(on_save_model_clicked)
load_model_button.on_click(on_load_model_clicked)
```

```python
# 9. Advanced Features
def plot_confusion_matrix(y_true, y_pred):
    from sklearn.metrics import confusion_matrix
    import seaborn as sns
    
    cm = confusion_matrix(y_true, y_pred)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=range(10), yticklabels=range(10))
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
plot_confusion_matrix(y_true, y_pred)
```
